```python
def month_profit_trace(profit, reach, make_figs=True): 
    if (profit.shape == (4,)) and (revenue.shape == (4,)): 
        week1 = np.array((reach[0], profit[0]))     #  
        week2 = np.array((reach[1],profit[1]))      #
        week3 = np.array((reach[2],profit[2]))      #DATA
        week4 = np.array((reach[3],profit[3]))      #
        
        week_total = week1+week2+week3+week4                         #this block of code is responsible for 
        week_performance = np.linalg.norm(reach*reach+profit*profit) #calculatating the given data 
        alpha = np.arctan(profit/reach)                              #
        reach_gradient = np.gradient(reach/profit)                   #
        

        plt.figure(figsize=(10,7))    #responsible for size of the chart
        plt.title('Bebang\'s Month Post Efficiency')
        plt.xlim(0,1.01*np.sum(reach))                            #limit of the x-axis and 
        plt.ylim(-np.sum(np.abs(profit)),np.sum(np.abs(profit)))  #y-axis in the chart 
        plt.xlabel('FB Post Reach Increment')  #lable of x-axis
        plt.ylabel('Profit')                   #and y-axis
        plt.grid()        #grid lines in the chart 
        n = 2
        
        plt.quiver(0,0, week1[0], week1[1],                                                      #
                   angles='xy', scale_units='xy',scale=1, color='yellowgreen', width=0.0025,     #
                   label='Week 1: {:.2f}'.format(np.linalg.norm(week1)))                         #
        plt.quiver(0,0, week2[0], week2[1],                                                      #this block of code is 
                   angles='xy', scale_units='xy',scale=1, color='blue', width=0.0025,            #responsible for the plot 
                   label='Week 2: {:.2f}'.format(np.linalg.norm(week2)))                         #and the characteristics 
        plt.quiver(0,0, week3[0], week3[1],                                                      #the each ploted dat 
                   angles='xy', scale_units='xy',scale=1, color='orange', width=0.0025,          #including the information 
                   label='Week 3: {:.2f}'.format(np.linalg.norm(week3)))                         #that will be shown at the 
        plt.quiver(0,0, week4[0], week4[1],                                                      #legend box 
                   angles='xy', scale_units='xy',scale=1, color='black', width=0.0025,           #
                   label='Week 4: {:.2f}'.format(np.linalg.norm(week4)))                         #
        
        

        plt.quiver(0,0, week_total[0], week_total[1],                                            #plot 
                   angles='xy', scale_units='xy',scale=1, color='red', width=0.005,              #for 
                  label='Efficiency: {:.2f} @ {}'.format(week_performance, reach_gradient))      #efficiency

        plt.legend(loc= 'upper left')                    #display legend and it's location

        if make_figs:                                                                               
            plt.savefig(f'LinAlg-Lab2-Bebang Post Eff-{(week_performance)}@{(reach_gradient)}.png', dpi=300)

        plt.show() 
    
    else:
       print('Dimension error')
```
